```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect("stores_1.db")
```

```python
q = """
    SELECT * --columns
    FROM ORDERS --Tables
    LIMIT 5 --No of rows
"""
pd.read_sql(q,conn)
```

```python
#join orders and orderdetails
q = """
   SELECT * FROM Orders O
   INNER JOIN orderdetails od
   on o.ordernumber = od.ordernumber
   LIMIT 5
   """
pd.read_sql(q,conn)
```

```python
#one column from each table
q = """
   SELECT customernumber , priceeach FROM Orders O
   INNER JOIN orderdetails od
   on o.ordernumber = od.ordernumber
   LIMIT 5
   """
pd.read_sql(q,conn)
```

```python
#join product table
q = """
    select * from orders o
    inner join orderdetails od
    on od.ordernumber = o.ordernumber
    join products p
    on od.productcode = p.productcode
    limit 5
    """

pd.read_sql(q,conn)
```

```python
#one column from each table
q = """
    select customernumber ,priceeach,buyprice 
    from orders o
    inner join orderdetails od
    on od.ordernumber = o.ordernumber
    join products p
    on od.productcode = p.productcode
    limit 5
    """

pd.read_sql(q,conn)
```

```python
#customer profitmargin
q = """
    select customernumber , (priceeach-buyprice) profit_margin 
    from orders o
    inner join orderdetails od
    on od.ordernumber = o.ordernumber
    join products p
    on od.productcode = p.productcode
    limit 5
    """

pd.read_sql(q,conn)
```

```python
#add column quantityordered
q = """
    select customernumber , (priceeach-buyprice) profit_margin , quantityordered
    from orders o
    inner join orderdetails od
    on od.ordernumber = o.ordernumber
    join products p
    on od.productcode = p.productcode
    limit 5
    """

pd.read_sql(q,conn)
```

```python
#profit calculation
q = """
    select customernumber , (priceeach-buyprice)*quantityordered profit
    from orders o
    inner join orderdetails od
    on od.ordernumber = o.ordernumber
    join products p
    on od.productcode = p.productcode
    """

pd.read_sql(q,conn)
```

```python
q = """
    select customernumber , sum((priceeach-buyprice)*quantityordered) total_profit
    from orders o
    inner join orderdetails od
    on od.ordernumber = o.ordernumber
    join products p
    on od.productcode = p.productcode
    GROUP BY 1
    LIMIT 10
    """

pd.read_sql(q,conn)
```

```python
#storing  total_profit by customersin a temporary table -->with clause
q = """
    WITH
    profit_by_customers AS
                      (
                      select customernumber , sum((priceeach-buyprice)*quantityordered) total_profit
                      from orders o
                      inner join orderdetails od
                      on od.ordernumber = o.ordernumber
                      join products p
                      on od.productcode = p.productcode
                      GROUP BY 1
                      )
     SELECT * FROM profit_by_customers
     """
pd.read_sql(q,conn)
```

```python
#join profit_by_customer (temporary table)
q = """
    WITH
    profit_by_customers AS
                      (
                      select customernumber , sum((priceeach-buyprice)*quantityordered) total_profit
                      FROM Orders O
                      INNER JOIN orderdetails od
                      on od.ordernumber = o.ordernumber
                      JOIN products p
                      on od.productcode = p.productcode
                      GROUP BY 1
                      )
     SELECT * FROM customers c
     JOIN profit_by_customers pc
     ON pc.customernumber = c.customernumber
     """
pd.read_sql(q,conn)
```

```python
#customer data
q = """
    WITH
    profit_by_customers AS
                      (
                      select customernumber , sum((priceeach-buyprice)*quantityordered) total_profit
                      FROM Orders O
                      INNER JOIN orderdetails od
                      on od.ordernumber = o.ordernumber
                      JOIN products p
                      on od.productcode = p.productcode
                      GROUP BY 1
                      )
     SELECT contactLastName, contactFirstName, City, Country, total_profit 
     FROM customers c
     JOIN profit_by_customers pc
     ON pc.customernumber = c.customernumber
     """
pd.read_sql(q,conn)
```

```python
#top 5 VIP customers: bring in the most profit for the customer
q = """
    WITH
    profit_by_customers AS
                      (
                      select customernumber , sum((priceeach-buyprice)*quantityordered) total_profit
                      FROM Orders O
                      INNER JOIN orderdetails od
                      on od.ordernumber = o.ordernumber
                      JOIN products p
                      on od.productcode = p.productcode
                      GROUP BY 1
                      )
     SELECT contactLastName, contactFirstName, City, Country, total_profit 
     FROM customers c
     JOIN profit_by_customers pc
     ON pc.customernumber = c.customernumber
     ORDER BY total_profit Desc
     LIMIT 5
     """
pd.read_sql(q,conn)
```
